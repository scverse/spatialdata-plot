```python
# Load via squidpy. visium_hne_sdata ships in-memory; persist once to a local
# zarr so write_element has a backing store for the persistence prototype.
# import squidpy as sq

SANDBOX_ZARR = Path.home() / "lustre" / "projects" / "spatialdata-plot" / "sandbox_data" / "visium_hne.zarr"
SANDBOX_ZARR.parent.mkdir(parents=True, exist_ok=True)

if SANDBOX_ZARR.exists():
    sdata = sd.read_zarr(SANDBOX_ZARR)
    print(f"loaded existing zarr: {SANDBOX_ZARR}")
else:
    sdata = sq.datasets.visium_hne_sdata()
    sdata.write(SANDBOX_ZARR)
    sdata = sd.read_zarr(SANDBOX_ZARR)
    print(f"wrote and reopened: {SANDBOX_ZARR}")

print("coordinate systems:", sdata.coordinate_systems)
print("images:", list(sdata.images.keys()))
```

```python
sdata
```

```python
(
    sdata
    .pl.render_images(element="hne")
    .pl.annotate(coordinate_systems="global")
)
```

```python
# Inspect what was saved by the annotate() session.
print("sdata.shapes keys:", list(sdata.shapes.keys()))
for k, s in sdata.shapes.items():
    print(f"  {k}: {len(s)} feature(s), bounds={s.total_bounds}")
```
